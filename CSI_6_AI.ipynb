## 3) <b>‚úÖ Control flow and functions</b>

<div class="box info">
  <div class="boxtitle"><span class="badge b-info">üß†</span> Control flow</div>
  <p><b>Control flow</b> is how Python decides <i>which statements run</i> and <i>how many times</i> they run.</p>
  <ul>
    <li>
      <b>If / elif / else</b> chooses exactly one branch based on a boolean condition (<code>True</code>/<code>False</code>).
      Conditions often use comparisons (<code>==</code>, <code>!=</code>, <code>&lt;</code>, <code>&gt;=</code>) and logical operators (<code>and</code>, <code>or</code>, <code>not</code>).
    </li>
    <li>
      <b>For loops</b> iterate through a sequence (e.g., a list or <code>range</code>).
      Use <code>range(n)</code> for ‚Äúrepeat n times‚Äù, and <code>enumerate(...)</code> when you want index + value.
    </li>
    <li>
      <b>While loops</b> repeat while a condition stays true. Be careful, if the condition never becomes false, you create an infinite loop.
      Use <code>break</code> to stop early, and <code>continue</code> to skip to the next iteration.
    </li>
    <li>
      <b>Indentation is syntax</b>. A block is defined by indentation, not braces.
      If indentation is wrong, your code still runs but your logic is different (often silently).
    </li>
  </ul>
</div>

<div class="box">
  <div class="boxtitle"><span class="badge b-info">üß©</span> Functions (why they matter)</div>
  <ul>
    <li><b>Functions</b> package code into reusable units, so you avoid copy-pasting and reduce mistakes.</li>
    <li><b>Inputs</b> are <b>parameters</b>, outputs come from <code>return</code>. If you do not return a value, the function returns <code>None</code>.</li>
    <li>Use <code>def name(...):</code> to define a function, then call it with parentheses, e.g., <code>name(x)</code>.</li>
    <li>Good practice, choose meaningful names, keep functions small, and test with a few simple examples.</li>
  </ul>
</div>



````python
score = 72  # Example mark (out of 100)

# Classify the score using if/elif/else (only one branch runs)
if score >= 70:
    label = "Distinction"
elif score >= 40:
    label = "Pass"
else:
    label = "Fail"

print("score:", score, "->", label)  # Print the result

count_70_plus = 0  # Counter for how many marks are 70 or above

# Loop through a list/array called `marks` and count distinctions
for m in marks:
    if m >= 70:
        count_70_plus += 1  # Increment counter when condition is met

print("Number of marks >= 70:", count_70_plus)  # Report the count
````

<div class="box">
  <div class="boxtitle"><span class="badge b-info">üß©</span> Functions</div>

  <p>
    A <b>function</b> packages a set of steps into a reusable unit. This improves readability, makes your code easier to test,
    and reduces errors caused by copy-paste.
  </p>

  <ul>
    <li><b>Define</b> a function with <code>def</code>, then <b>call</b> it using parentheses, e.g., <code>my_func(x)</code>.</li>
    <li><b>Inputs</b> are called <b>parameters</b> (or arguments when calling). They let you reuse the same logic on different data.</li>
    <li><b>Outputs</b> come from <code>return</code>. If a function does not return anything explicitly, it returns <code>None</code>.</li>
    <li><b>Good practice</b>, keep functions small, give them clear names, and test them with a few simple cases.</li>
  </ul>
</div>


````python
def grade_label(mark: int) -> str:
    """Return a grade label based on threshold rules (70+, 40‚Äì69, below 40)."""
    # If the mark is 70 or above, classify as Distinction
    if mark >= 70:
        return "Distinction"
    # Otherwise, if the mark is 40 or above, classify as Pass
    elif mark >= 40:
        return "Pass"
    # Otherwise (below 40), classify as Fail
    else:
        return "Fail"

# Test the function on a few example marks
for t in [75, 55, 20]:
    print(t, "->", grade_label(t))  # Print each mark and its computed label
````

def count_in_range(values, low, high):
    count = 0
    for v in values:
        if low <= v <= high:
            count += 1
    return count

print("Marks in [60, 79] =", result)

## 4) <b>‚úÖ Pandas for tabular data</b>

<div class="box info">
  <div class="boxtitle"><span class="badge b-info">üìä</span> What you will practise</div>
  <ul>
    <li><b>Create a DataFrame</b>, think of it as an Excel-like table with rows and named columns.</li>
    <li><b>Inspect it</b> so you understand the data before you analyse it,
      <code>head()</code> shows a quick preview, <code>info()</code> shows column types and missing values, <code>describe()</code> gives summary statistics.</li>
    <li><b>Filter rows and create new columns</b>, select subsets using conditions (e.g., ‚Äúmark ‚â• 70‚Äù), and compute derived columns (e.g., pass/fail labels).</li>
    <li><b>Handle missing values sensibly</b>, first detect them, then decide whether to remove rows or fill values (only when it makes sense for the problem).</li>
  </ul>
</div>


````python
# Create a small dataset (no internet dependency)
df = pd.DataFrame({
    "student": ["Aisha", "Ben", "Chen", "Dina", "Eren", "Fatima", "George", "Hana"],
    "hours_studied": [2.0, 5.5, 3.0, 8.0, np.nan, 6.0, 1.5, 7.0],
    "attendance": [0.82, 0.91, 0.76, 0.95, 0.88, np.nan, 0.70, 0.93],
})

# Generate random noise (mean=0, std=4) to make the synthetic scores realistic rather than perfectly linear
noise = np.random.normal(loc=0, scale=4.0, size=len(df))

# Create a synthetic "score" using a simple linear rule:
# baseline (40) + 5 points per study hour + random noise

# Note: we temporarily replace missing hours_studied values with the column mean to avoid NaNs in the score
df["score"] = np.round(
    40 + 5 * df["hours_studied"].fillna(df["hours_studied"].mean()) + noise,
    1  # round to 1 decimal place for tidy presentation
)

# Display the updated DataFrame
df
````

````python
# df.shape -> (rows, columns): quick size check
print("Shape:", df.shape)

# df.head() -> first few rows: checks values loaded correctly
display(df.head())

# df.info() -> column names, non-null counts, dtypes: good for missing values + type issues
print("\nInfo:")
df.info()

# describe() on numeric columns only: summary stats (mean, std, quartiles, etc.)
# we use select_dtypes to be compatible with Pandas versions
print("\nDescribe (numeric columns):")
display(df.select_dtypes(include="number").describe())
````

````python
# Handle missing values
# Strategy: for each numeric column, replace missing entries (NaN) with the column mean
# Rationale: this is a simple baseline imputation method that preserves the dataset size

# Make a copy so the original df remains unchanged
df_clean = df.copy()

# Impute missing values in selected numeric columns
for col in ["hours_studied", "attendance"]:
    # Compute the mean of the column (ignoring NaNs by default)
    col_mean = df_clean[col].mean()
    # Fill NaNs with the mean value
    df_clean[col] = df_clean[col].fillna(col_mean)

# Display the cleaned DataFrame
df_clean
````

## 5) <b>‚úÖ Visualisation with Matplotlib</b>

<div class="box info">
  <div class="boxtitle"><span class="badge b-info">üìà</span> Three plots you should be able to produce</div>
  <ul>
    <li>
      <b>Histogram</b>, shows the <b>distribution</b> of one variable by grouping values into ‚Äúbins‚Äù.
      Use it to check spread, typical values, skewness, and potential outliers.
    </li>
    <li>
      <b>Scatter plot</b>, shows the <b>relationship</b> between two variables (each point is one observation).
      Use it to spot trends (positive/negative), clusters, and outliers.
    </li>
    <li>
      <b>Line plot</b>, use when data has a natural <b>order</b> (for example, time or iterations).
      Use it to see changes over time, peaks/troughs, and overall patterns.
    </li>
  </ul>
  <p><b>Good practice:</b> always add a title, axis labels, and choose sensible scales so the plot is readable.</p>
</div>



````python
# Histogram: visualise the distribution of scores
# Purpose: check how scores are spread (centre, variability) and whether any values look unusual

# Create a new figure with a controlled size (width=6 inches, height=4 inches)
plt.figure(figsize=(6,4))

# Plot a histogram of the 'score' column
# bins=7 groups the scores into 7 intervals (a simple choice for a small dataset)
plt.hist(df_clean["score"], bins=7)

# Add a clear title and axis labels for interpretation
plt.title("Score distribution")
plt.xlabel("Score")
plt.ylabel("Count")

# Add a light grid to improve readability without overwhelming the plot
plt.grid(True, alpha=0.25)

# Render the plot
plt.show()
````

````python
# Scatter plot: visualise the relationship between hours studied and score
# Purpose: check whether higher study time tends to be associated with higher scores (a possible linear trend)

# Create a new figure with a controlled size (width=6 inches, height=4 inches)
plt.figure(figsize=(6,4))

# Plot each student as a point: x-axis = hours studied, y-axis = score
plt.scatter(df_clean["hours_studied"], df_clean["score"])

# Add a clear title and axis labels for interpretation
plt.title("Hours studied vs score")
plt.xlabel("Hours studied")
plt.ylabel("Score")

# Add a light grid to make reading values easier
plt.grid(True, alpha=0.25)

# Render the plot
plt.show()
````

```python
# Step 0: work on a copy to avoid accidental edits to df_clean
df_hw = df_clean.copy()

# Step 1: performance_index
# - score is already on a 0‚Äì100-ish scale
# - attendance is on 0‚Äì1, so multiply by 100 before combining
df_hw["performance_index"] = 0.6 * df_hw["score"] + 0.4 * (100 * df_hw["attendance"])

# Step 2: performance_band (High / Medium / Low)
# Use np.select for clean, vectorised multi-condition assignment
conditions = [
    df_hw["performance_index"] >= 75,
    (df_hw["performance_index"] >= 60) & (df_hw["performance_index"] < 75),
    df_hw["performance_index"] < 60
]
choices = ["High", "Medium", "Low"]

df_hw["performance_band"] = np.select(conditions, choices, default="Medium")

# Step 3: top 3 students by performance_index
top3 = (
    df_hw.sort_values("performance_index", ascending=False)
         .loc[:, ["student", "performance_index", "performance_band"]]
         .head(3)
)
print("Top 3 students (by performance_index):")
display(top3)

# Step 4: group summary (mean performance_index per band)
summary = (
    df_hw.groupby("performance_band")["performance_index"]
         .mean()
         .sort_values(ascending=False)
)
print("Mean performance_index by band:")
display(summary)

# Step 5: scatter plot, hours_studied vs performance_index, coloured by performance_band
plt.figure(figsize=(6,4))

for band, sub in df_hw.groupby("performance_band"):
    plt.scatter(sub["hours_studied"], sub["performance_index"], label=band)

plt.title("Hours studied vs performance index (coloured by band)")
plt.xlabel("Hours studied")
plt.ylabel("Performance index")
plt.grid(True, alpha=0.25)
plt.legend(title="Performance band")
plt.show()
