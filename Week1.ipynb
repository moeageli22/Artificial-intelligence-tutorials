# EXERCISE 2: Summary statistics from a list

```python
print("\n\n" + "=" * 70)
print("EXERCISE 2: Summary statistics from a list")
print("=" * 70)

# Solution
avg_mark = sum(marks) / len(marks)
min_mark = min(marks)
max_mark = max(marks)

print(f"Average = {avg_mark:.1f}, Min = {min_mark}, Max = {max_mark}.")

print("\n Exercise 2 complete!")
print("Explanation:")
print("- sum() adds all values, len() counts them → average")
print("- min() and max() are built-in functions")
print("- {avg_mark:.1f} formats to 1 decimal place")
```

# EXERCISE 3: Function + loop

```python
print("\n\n" + "=" * 70)
print("EXERCISE 3: Function + loop")
print("=" * 70)

# Solution
def count_in_range(values, low, high):
    """
    Count how many values fall in the inclusive range [low, high].

    Args:
        values: list of numbers
        low: minimum value (inclusive)
        high: maximum value (inclusive)

    Returns:
        int: count of values in range
    """
    count = 0
    for v in values:
        if low <= v <= high:
            count += 1
    return count

# Apply the function
result = count_in_range(marks, 60, 79)
print(f"Number of marks in [60, 79]: {result}")

print("\n Exercise 3 complete!")
print("Explanation:")
print("- We iterate through every value in the list")
print("- The condition 'low <= v <= high' checks if v is in range (inclusive)")
print("- We increment counter when condition is True")
print("- Function can be reused with different ranges")
```

# DATA SETUP FOR EXERCISE 4 & 5

```python
print("\n\n" + "=" * 70)
print("Creating DataFrame for Exercises 4 & 5...")
print("=" * 70)

# Create a small dataset
df = pd.DataFrame({
    "student": ["Aisha", "Ben", "Chen", "Dina", "Eren", "Fatima", "George", "Hana"],
    "hours_studied": [2.0, 5.5, 3.0, 8.0, np.nan, 6.0, 1.5, 7.0],
    "attendance": [0.82, 0.91, 0.76, 0.95, 0.88, np.nan, 0.70, 0.93],
})

# Generate random noise
noise = np.random.normal(loc=0, scale=4.0, size=len(df))

# Create synthetic score
df["score"] = np.round(
    40 + 5 * df["hours_studied"].fillna(df["hours_studied"].mean()) + noise,
    1
)

print("\nOriginal DataFrame:")
print(df)

# Handle missing values
df_clean = df.copy()

for col in ["hours_studied", "attendance"]:
    col_mean = df_clean[col].mean()
    df_clean[col] = df_clean[col].fillna(col_mean)

print("\nCleaned DataFrame (missing values filled with mean):")
print(df_clean)
```

# EXERCISE 4: Filtering and new columns

```python
print("\n\n" + "=" * 70)
print("EXERCISE 4: Filtering and new columns")
print("=" * 70)

# Solution
df_tmp = df_clean.copy()

# 1) Create score_band column
df_tmp["score_band"] = np.where(df_tmp["score"] >= 70, "Distinction", "Pass")

# 2) Filter for students with ≥5 hours studied AND ≥0.85 attendance
condition = (df_tmp["hours_studied"] >= 5) & (df_tmp["attendance"] >= 0.85)
df_focus = df_tmp[condition]

# 3) Display selected columns
cols = ["student", "hours_studied", "attendance", "score", "score_band"]
print("\nFiltered students (hours_studied ≥ 5 AND attendance ≥ 0.85):")
print(df_focus[cols])

print("\n Exercise 4 complete!")
print("Explanation:")
print("- np.where(condition, value_if_true, value_if_false) creates new column")
print("- Use & for AND, | for OR when combining conditions")
print("- Use parentheses around each condition when combining")
print("- df[columns_list] selects specific columns")
```

# EXERCISE 5: Make a clear plot

```python
# Histogram: visualise the distribution of scores
# Purpose: check how scores are spread (centre, variability) and whether any values look unusual

# Create a new figure with a controlled size (width=6 inches, height=4 inches)
plt.figure(figsize=(6,4))

# Plot a histogram of the 'score' column
# bins=7 groups the scores into 7 intervals (a simple choice for a small dataset)
plt.hist(df_clean["score"], bins=7)

# Add a clear title and axis labels for interpretation
plt.title("Score distribution")
plt.xlabel("Score")
plt.ylabel("Count")

# Add a light grid to improve readability without overwhelming the plot
plt.grid(True, alpha=0.25)

# Render the plot
plt.show()
```

```python
# Scatter plot: visualise the relationship between hours studied and score
# Purpose: check whether higher study time tends to be associated with higher scores (a possible linear trend)

# Create a new figure with a controlled size (width=6 inches, height=4 inches)
plt.figure(figsize=(6,4))

# Plot each student as a point: x-axis = hours studied, y-axis = score
plt.scatter(df_clean["hours_studied"], df_clean["score"])

# Add a clear title and axis labels for interpretation
plt.title("Hours studied vs score")
plt.xlabel("Hours studied")
plt.ylabel("Score")

# Add a light grid to make reading values easier
plt.grid(True, alpha=0.25)

# Render the plot
plt.show()
```
